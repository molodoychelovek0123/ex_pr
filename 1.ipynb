```python
import pandas
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
import keras
from matplotlib import pyplot as plt
```

Датасет ВИН - https://archive.ics.uci.edu/dataset/109/wine

```python
dataframe = pandas.read_csv("iris.csv", header=None)
dataset = dataframe.values
X = dataset[:,0:4].astype(float)
Y = dataset[:,4]
```

```python
encoder = LabelEncoder()
encoder.fit(Y)
encoded_Y = encoder.transform(Y)
dummy_y = to_categorical(encoded_Y)
```

```python
model = Sequential()
model.add(Dense(4, activation='relu'))
model.add(Dense(3, activation='softmax'))

model2 = Sequential()
model2.add(Dense(4, activation='relu'))
model2.add(Dense(3, activation='softmax'))

model3 = Sequential()
model3.add(Dense(4, activation='relu'))
model3.add(Dense(3, activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy', metrics=['accuracy'])
model2.compile(optimizer='adam',loss='categorical_crossentropy', metrics=['accuracy'])
model3.compile(optimizer='SGD',loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(X, dummy_y, epochs=75, batch_size=10, validation_split=0.1)
history2 = model2.fit(X, dummy_y, epochs=105, batch_size=16, validation_split=0.1)
history3 = model3.fit(X, dummy_y, epochs=105, batch_size=16, validation_split=0.1)
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Исходная', color='blue')
plt.plot(history2.history['loss'], label='Измененная №1', color='green')
plt.plot(history3.history['loss'], label='Измененная №2', color='pink')
plt.xlabel('Эпохи')
plt.ylabel('Потери')

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Исходная', color='blue')
plt.plot(history2.history['accuracy'], label='Измененная №1', color='green')
plt.plot(history3.history['accuracy'], label='Измененная №2', color='pink')
plt.xlabel('Эпохи')
plt.ylabel('Точность')
plt.legend(bbox_to_anchor=( 1.02 , 1 ), loc='upper left', borderaxespad= 0 )
plt.show()
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['val_loss'], label='Исходная', color='blue')
plt.plot(history2.history['val_loss'], label='Измененная №1', color='green')
plt.plot(history3.history['val_loss'], label='Измененная №2', color='pink')
plt.xlabel('Эпохи')
plt.ylabel('Потери')

plt.subplot(1, 2, 2)
plt.plot(history.history['val_accuracy'], label='Исходная', color='blue')
plt.plot(history2.history['val_accuracy'], label='Измененная №1', color='green')
plt.plot(history3.history['val_accuracy'], label='Измененная №2', color='pink')
plt.xlabel('Эпохи')
plt.ylabel('Точность')
plt.legend(bbox_to_anchor=( 1.02 , 1 ), loc='upper left', borderaxespad= 0 )
plt.show()
```
